```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from spectral import SpectralClustering
import sklearn
```

```python
X,y = make_moons(n_samples = 150, noise = 0.07, random_state = 21)

spectral = SpectralClustering(n_clusters = 2, random_state = 42)

y_pred = spectral.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c = y_pred)
plt.show()
```

```python
clustering = sklearn.cluster.SpectralClustering(n_clusters= 2, random_state = 42).fit(X)
plt.scatter(X[:,0], X[:,1], c = clustering.labels_)
plt.show()
```
